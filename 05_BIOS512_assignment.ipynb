```r
library('tidyverse')
library('dplyr')
```

```r
wind = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv', 
                col_types = cols())
```

```r
wind |> head()
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
province_territory = wind |> pull(province_territory) |> factor()
```

```r
province_territory |> levels()
```

```r
province_sorted = province_territory |> fct_infreq() |> fct_rev()
```

```r
province_sorted |> levels()
```

```r
p = ggplot(wind, aes(y = province_sorted)) + 
        geom_bar()

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
manufacturer = wind |> pull(manufacturer) |> factor()
```

```r
manufacturer |> levels()
```

```r
manufacturer_sorted = manufacturer |> fct_infreq() |> fct_lump(n = 6) |> fct_rev()
```

```r
manufacturer_sorted |> levels()
```

```r
manufacturer_sorted |> table()
```

```r
p = ggplot(wind, aes(y = manufacturer_sorted)) + 
        geom_bar()

p
```
