Use `geom_bar` to make a histogram of penguin counts by `year`.

```r
options(repr.plot.height = 3, repr.plot.width = 4)

p = ggplot(penguins, aes(x = year)) + geom_bar(fill = 'steelblue', color = 'steelblue', alpha = 0.75) +
    labs(x = 'Year', y = 'Number of Peguins', title = 'Count of Penguins by Year')+
    theme_minimal() + theme(plot.title = element_text(size = 9, hjust = .5)) 
p
```

Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.height = 5, repr.plot.width = 5)
p_mass = ggplot(penguins, aes(x = body_mass_g)) + geom_histogram(fill = 'steelblue', color = 'steelblue', alpha = 0.75, binwidth = 100, na.rm = TRUE) +
    facet_wrap(~year, ncol = 1) + scale_x_continuous(labels = comma_format()) +
    theme_minimal() + theme(plot.title = element_text(size = 9), strip.text.x = element_text(face = 'bold', hjust = 1, color = '#63666A', size = 10),
                           plot.margin = margin(t = 0, r = 1, b = 0, l = 0, unit = "cm"))  +
    labs(y = 'Number of Penguins', x = 'Body Mass in Grams') + labs(title = "Histogram of Penguin Body Mass for Measurements taken from 2007-2009")
p_mass
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

- Male Penguins are often larger than Female penguins
    - Male and Female Chinstrap penguis are closer in size compared to other penguin species

```r
options(repr.plot.height = 5, repr.plot.width = 4)
p_scat = ggplot(penguins, aes(y = sex, x = body_mass_g)) + geom_point(na.rm = TRUE, position = position_jitter(height = 0.2, width = 0), size = 1, alpha = 0.5, color = '#63666A') +
    facet_wrap(~species, ncol = 1) + theme_minimal() + theme(strip.text.x = element_text(face = 'bold', hjust = 0, color = '#63666A', size = 10)) +
    labs(x = 'Body Mass in Grams') + scale_x_continuous(labels = comma_format())

p_scat
```
